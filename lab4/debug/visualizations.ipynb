```python
import json
import matplotlib.pyplot as plt
import os
```

```python
epochs_dir = "./saved_model/logs/"
```

```python
epoc_files = os.listdir(epochs_dir)
```

```python
metrics = []
for file in epoc_files:
    with open(epochs_dir + file, 'r', encoding='utf-8') as fd:
        data = json.loads(fd.read())
    metrics.append(data)
```

```python
metrics = metrics[1:]
```

```python
x = list(range(len(metrics)))
y1 = list(map(lambda v: v['val_loss'], metrics))
y2 = list(map(lambda v: v['train_loss'], metrics))
```

```python
metrics
```

```python
plt.figure(figsize=(5,4))
plt.plot(x, y1, label='eval')
plt.plot(x, y2, label='train')
plt.xlabel("Эпоха обучения")
plt.ylabel("Значение функции ошибки")
plt.legend()
plt.grid()
plt.show()
```
